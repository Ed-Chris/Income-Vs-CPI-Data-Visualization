CPI is defined with the following formula:

---

$CPI = (Current  Year / Base Year) * 100$

**where:**

**CPI**  = Consumer Price Index in Current Period

**Current** Period = Cost of Market Basket in Current Period

**Base** = Cost of Market Basket in Base Period**

```python
merged_pivot.rename(columns={'REF_DATE':'Date', 'GEO' : 'Geography'}, inplace=True)
merged_pivot
```

```python
# Using 2002 as base for Income Index as base year for cpi is 2002, this is to see the changes relative to a common reference point.
# creating a new column for income index and taking 2002 as base year, where the average income is converted to 100 for base year and the rest changes accordingly, for all provinces seperately. 
# Calculate Income Index for each Geography
def calculate_income_index(group):
    base_year_value = group['Average income (excluding zeros)'][group['Date'] == 2002].values[0]
    group['Income Index'] = (group['Average income (excluding zeros)'] / base_year_value) * 100
    return group

merged_pivot = merged_pivot.groupby('Geography').apply(calculate_income_index)

merged_pivot
```

```python
# Using 2002 as base for Income Index as base year for cpi is 2002, this is to see the changes relative to a common reference point.
# creating a new column for income index and taking 2002 as base year, where the average income is converted to 100 for base year and the rest changes accordingly, for all provinces seperately. 
# Calculate Income Index for each Geography
def calculate_median_income_index(group):
    base_year_value = group['Median income (excluding zeros)'][group['Date'] == 2002].values[0]
    group['Median Income Index'] = (group['Median income (excluding zeros)'] / base_year_value) * 100
    return group

merged_pivot = merged_pivot.groupby('Geography').apply(calculate_median_income_index)

merged_pivot
```

```python

# Collect data for Canada
canada_data = merged_pivot[merged_pivot["Geography"] == "Canada"]

# Collect data for the 8 CPI categories and date
select_columns = ["Date", "Food", "Shelter", "Household operations, furnishings and equipment", 
                  "Clothing and footwear", "Transportation", "Health and personal care", "Recreation, education and reading", 
                  "Alcoholic beverages, tobacco products and recreational cannabis"]

canada_data = canada_data[select_columns]

# Set the Date as index to plot our 8 categories
canada_data.set_index("Date", inplace = True)
canada_data
```

```python

# Create a 4 x 2 grid of subplots for each of the 8 categories
fig = sp.make_subplots(rows=4, cols=2, subplot_titles=canada_data.columns)

# Define colors for the traces
colors = px.colors.qualitative.Plotly

# Add traces for each category using a forloop and calculate row + column size
for i, category in enumerate(canada_data.columns):
    row = (i // 2) + 1  # Calculate the row for the subplot
    col = (i % 2) + 1   # Calculate the column for the subplot
    trace = go.Scatter(x=canada_data.index, y=canada_data[category], mode='lines', name=category, line=dict(color=colors[i]))
    fig.add_trace(trace, row=row, col=col)

# Update subplot layout, keep x-axis empty (don't want labels there)
fig.update_xaxes(title_text='Years', row = 4) 
fig.update_yaxes(range=[45, 155])
fig.update_yaxes(title_text='CPI Value', col=1) 

# Update subplot title font
for i, title in enumerate(canada_data.columns):
    fig.update_annotations(selector=dict(text=title), font=dict(size=11))  

# Update figure layout with title in center
fig.update_layout(
    title_text='CPI Categories Over Time in Canada',
    title_x=0.5, 
    showlegend=False,
    height=800,
    width=800
)

# Show the plot
fig.show()
```

```python
# Create a line graph with the CPI categories overtime all on the same graph
fig = px.line(
    canada_data,
    x=canada_data.index,
    y=canada_data.columns,
    labels={'index': 'Date', 'value': 'CPI Value'},
)

# Customize the layout of the graph
fig.update_layout(
    title='Consumer Price Index (CPI) Categories Over Time (Canada)',
    title_x=0.5,
    xaxis_title='Date',
    yaxis_title='CPI Value',
    showlegend=True,
    legend=dict(orientation='h', y=-0.15, x=0.35),  
    height=800,
    xaxis=dict(tickangle=-45, showline=True, linewidth=1, linecolor='lightgray'), 
    yaxis=dict(gridcolor='lightgray'),
    font=dict(family="Arial", size=12),
)
# Update the size of the lines
for trace in fig.data:
    trace.line.width = 3
    
# Show the plot
fig.show()
```

```python
merged_pivot['Affordability_Index'] = merged_pivot['Average income (excluding zeros)'] / merged_pivot['All-items']
merged_pivot['Median Affordability_Index'] = merged_pivot['Median income (excluding zeros)'] / merged_pivot['All-items']
merged_pivot1 = merged_pivot[merged_pivot['Geography'] != 'Canada']

# Calculating national average affordability index
national_average_affordability = merged_pivot1['Average income (excluding zeros)'].sum() / merged_pivot1['All-items'].sum()
national_median_average_affordability = merged_pivot1['Median income (excluding zeros)'].sum() / merged_pivot1['All-items'].sum()

display(merged_pivot1)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set the style of seaborn plots
sns.set(style="whitegrid")

# Create a figure with two subplots
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

# Plot the first barplot on the first subplot
sns.barplot(x='Geography', y='Affordability_Index', data=merged_pivot1, palette='viridis', ax=axes[0])
axes[0].axhline(y=national_average_affordability, color='r', linestyle='--', label='National Average Affordability')
axes[0].set_xlabel('Province')
axes[0].set_ylabel('Affordability Index')
axes[0].set_title('Affordability Index by Province')
axes[0].tick_params(axis='x', rotation=45)
axes[0].legend()

# Plot the second barplot on the second subplot
sns.barplot(x='Geography', y='Median Affordability_Index', data=merged_pivot1, palette='viridis', ax=axes[1])
axes[1].axhline(y=national_median_average_affordability, color='r', linestyle='--', label='National Median Average Affordability')
axes[1].set_xlabel('Province')
axes[1].set_ylabel('Median Affordability Index')
axes[1].set_title('Median Affordability Index by Province')
axes[1].tick_params(axis='x', rotation=45)
axes[1].legend()

# Ensure proper spacing between subplots
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import matplotlib.pyplot as plt

# Filter the merged_pivot DataFrame to include data for the year 1981
merged_pivot_1981 = merged_pivot1[merged_pivot1['Date'] == 1981]

# Load the geospatial data from the shapefile
gdf = gpd.read_file("C:/Users/17adi/OneDrive/Desktop/canadashape")

# Assuming 'Affordability_Index' is a column in your dataset and 'Geography' is a common column
# Merge the geospatial data with your dataset based on the common column ('Geography')
merged_data_1981 = gdf.merge(merged_pivot_1981, left_on='PRENAME', right_on='Geography', how='inner')

# Create the heatmap
fig, ax = plt.subplots(1, 1, figsize=(12, 8))
merged_data_1981.plot(column='Affordability_Index', cmap='viridis', linewidth=0.8, ax=ax, edgecolor='0.8', legend=True)

# Add rounded numbers on top of each province (rounded to two decimal places)
for x, y, label in zip(merged_data_1981.geometry.centroid.x, merged_data_1981.geometry.centroid.y, merged_data_1981['Affordability_Index']):
    ax.text(x, y, round(label, 2), fontsize=10, ha='center', va='center')


ax.set_title('Affordability Index Heatmap of 1981')
plt.axis('off')
plt.show()
```
